# Research Question

What impacts do regions, along with production methods (conventional vs. organic) have on U.S. avocado price fluctuations ranging from 2015 to 2023?



## Background and Prior Work



Avocados, often hailed as a superfruit, are packed with a wide array of vitamins and minerals. Owing to their extensive health benefits, they have sustained their popularity over the past decade, consistently featuring across various culinary presentations. The consumption of avocados have become significant over the past years, thus causing an increase in demand. Because of this, the price of avocados has been affected for a number of reasons. Factors such as the seasonality of avocado production, with specific periods favoring optimal growth, the type of avocado (organic or conventional), and the region of cultivation play a role in influencing avocado’s availability and quality.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Avocado prices fluctuate depending on the time of year. Tanmay Deshpande, the author of ‘Avocado Price Forecast,’ used Auto Regressive Integrated Moving Average (ARIMA) and Seasonal Auto Regressive Integrated Moving Average (SARIMA) model to find the correlation between the two types of avocado prices and the time of year. He discovered that the average prices of avocados fluctuate during a certain time of year as it would rise around September to November, but would drop again around December and January. This suggests that there may be more of a demand for avocados since they are seasonal fruits, which is why costs increase during a specific time of year in which they are produced.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

However, the costs of avocados also vary depending on the region of which they are produced and sold for a number of reasons. Mario Caesar, author of  ‘Avocado Price Regression w/ PyCaret & EDA,’ discusses how different areas around the United States play a big role in the contribution of avocado sales and prices, especially with the size of the bag of avocados. His article expresses how the West (excluding California) and California are more likely to purchase both conventional and organic avocados—more so conventional—in comparison to the other top five areas that purchase avocados—the Northeast, Great Lakes and South Central, which begs the question as to why there is such a difference in avocado consumption between different regions. Additionally, the type of avocado that is produced also affects cost and people’s decision to purchase as they are grown using different methods. Caesar concludes that conventional avocados that are within bags are more likely to be purchased in comparison to organic avocados since they tend to be cheaper and bigger in volume and size. Even though organic avocados are produced in a better environment, their prices are higher for that reason.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Despite how much work people had put into predicting the price of avocados, there are limitations to previous works. Although there are price predictions on the avocadoes, regional factors and planting methods had not been the main focus. For example, in the ‘Avocado Price Forecast,’ time of the year is considered. On the other hand, in ‘Forecasting Price Trends in the U.S. Avocado (Persea americana Mill.) Market’ by Evans and Nalampang, individuals’ disposable income and Hispanic immigration population is considered. However, production method and region are two factors that are relatively rarely discussed, which is our main focus.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Interestingly from further inspection, values under `Region` vary by city name, states, and entire regions of the United States (e.g. Midsouth). Thus, we decided to standardize this column by creating a function that hardcodes each city to be the state that it is in to make the data more representative in the form of US states. Afterwards, we remove the rows that contain informally-defined regions of the United States, since we cannot define what regions belong to what state. We renamed our `Region` column to `State` so that the label is consistent with the changes. 

```python
# Convert Regions into respective states
def name_changing(vals):
    
    if vals in ['California', 'SanFrancisco', 'SanDiego', 'LosAngeles', 'Sacramento']:
        return 'California'
    elif vals in ['Albany', 'BuffaloRochester', 'Syracuse']:
        return 'New York'
    elif vals in ['Seattle', 'Spokane']:
        return 'Washington'
    elif vals in ['Atlanta', 'Plains']:
        return 'Georgia'
    elif vals in ['Miami', 'Orlando', 'Tampa', 'MiamiFtLauderdale']:
        return 'Florida'
    elif vals in ['Wichita']:
        return 'Kansas'
    elif vals in ['Portland']:
        return 'Oregon'
    elif vals in ['PhoenixTucson']:
        return 'Arizona'
    elif vals in ['Boston']:
        return 'Massachusetts'
    elif vals in ['DallasFtWorth', 'Houston']:
        return 'Texas'
    elif vals in ['BirminghamMontgomery']:
        return 'Alabama'
    elif vals in ['Charlotte', 'RaleighGreensboro']:
        return 'North Carolina'
    elif vals in ['StLouis']:
        return 'Missouri'
    elif vals in ['HarrisburgScranton', 'Philadelphia', 'Pittsburgh']:
        return 'Pennsylvania'
    elif vals in ['CincinnatiDayton', 'Columbus', 'Toledo']:
        return 'Ohio'
    elif vals in ['Indianapolis']:
        return 'Indiana'
    elif vals in ['Detroit', 'GrandRapids', 'GreatLakes']:
        return 'Michigan'
    elif vals in ['Chicago', 'PeoriaSpringfield']:
        return 'Illinois'
    elif vals in ['Baltimore', 'BaltimoreWashington']:
        return 'Maryland'
    elif vals in ['Providence']:
        return 'Rhode Island'
    elif vals in ['Denver']:
        return "Colorado"
    elif vals in ['HartfordSpringfield']:
        return 'Connecticut'
    elif vals in ['LasVegas']:
        return 'Nevada'
    elif vals in ['Louisville']:
        return 'Kentucky'
    elif vals in ['Nashville']:
        return 'Tennessee'
    elif vals in ['NewOrleans']:
        return 'Louisiana'
    elif vals in ['RichmondNorfolk', 'Roanoke']:
        return 'Virginia'
    elif vals in ['SouthCarolina']:
        return 'South Carolina'
    else:
        return 'Idaho'  #boise
    
```

```python
# Delete informally-defined regions 
    #Northeast, Midsouth, NorthernNewEnglish as it is neither a state or city; data is unspecified
    #SouthCentral, not sure if it means city in LA or the south central state; data is unclear
    #TotalUS and WestUS; data is unclear
    #WestTexNewMexico; data is unclear
    #Jacksonville appears in different states; unable to specify data
values_to_delete = ['Midsouth', 'SouthCentral', 'WestTexNewMexico', 'Southeast', 'TotalUS', 'Jacksonville', 'Northeast']
df = df[~df['Region'].isin(values_to_delete)]
```
